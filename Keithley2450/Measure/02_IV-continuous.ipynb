```python
import pySMUuvic as psu

# # To check Keithley Address
# psu.get_Keithley_address()

save_dir = '/Users/lznus/Desktop/Data-Keithley/'
address = 'USB0::0x05E6::0x2450::04491080::INSTR'
operation_mode = '02_IV'

project_dir = psu.project_start(save_dir, operation_mode)
```

```python
#### Run the measurement
psu.run_IV(
    project_dir = project_dir,
    v_range = [0, 10], #V
    step_size = 0.5, #V/step
    scan_rate = 10, #V/s
    ILIMIT = 0.05, # A
    direction = 'B', # B:both (F -> R), F:forward, R:reverse
    terminals = 'REAR', # 'FRONT' or 'REAR' panel
    address = address
)
```

```python
# Grab and show the data list
data_list = psu.data_list(project_dir)
```

```python
plot_index = [7,9] # Chose the numbers

psu.IV_plot(data_list, plot_index, project_dir, xlog = False, ylog = False)
```
