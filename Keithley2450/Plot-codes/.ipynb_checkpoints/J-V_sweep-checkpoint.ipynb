```python
"""
Combine J-V sweep at Forward and Reverse bias
"""

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import glob

data_dir = "/Users/yukiharuta/Desktop/Temp-Data/Doping-for-pn/Keithley/240214_Bi-doped_MAPbBr3_IV"

csv_list = glob.glob(f'{data_dir}/*.csv')
csv_list.sort()
```

```python
FR_classify = ['negative', 'positive'] # key word to specify if it is FORWARD range or REVERSE range
scan_direction = ['-F','-R'] # key word to specify if it is FORWARD scan or REVERSE scan
log_plot = True

previous_sample_id = ''
last_data = False

sample_id_list = []
for csv in csv_list:
    filename = os.path.basename(csv)[:-4]
    fileinfo = filename.split('_')
    sample_id = = '_'.join(fileinfo[:-1])
    sample_id_list.append(sample_id)
    
for i, csv in enumerate(csv_list):
    filename = os.path.basename(csv)[:-4]
    fileinfo = filename.split('_')
    sample_id = = '_'.join(fileinfo[:-1])
    
    if sample_id != previous_sample_id:
        first_data = True
        
    if first_data:
        print(f'\n{sample_id}')
        first_data = False
    
    df = pd.read_csv(csv)
    X = df['Voltage (V)'].to_list()
    Y = df['Current (A)'].to_list()
    
    if log_plot:
        Y = [abs(n) for n in Y]
    
    # plot forward bias
    if FR_classify[0] in filename:
        if scan_direction[0] in filename:
            plt.plot(X,Y,c='blue',label='forward',marker='o')
        elif scan_direction[1] in filename:
            plt.plot(X,Y,c='green',label='reverse',marker='o')
    
    # plot negative bias
    elif FR_classify[1] in filename:
        X = [-n for n in X]
        if log_plot:
            Y = [abs(n) for n in Y]
        else:
            Y = [-n for n in Y]
        if scan_direction[0] in filename:
            plt.plot(X,Y,c='blue',label='forward',marker='o')
        elif scan_direction[1] in filename:
            plt.plot(X,Y,c='green',label='reverse',marker='o')
    
    if csv == csv_list[-1]:
        first_data = True
        if log_plot:
            plt.yscale('log')
        plt.xlabel('Voltage (V)')
        plt.ylabel('Current (A)')
        plt.show()

    elif sample_id != sample_id_list[i+1]:
        first_data = True
        if log_plot:
            plt.yscale('log')
        plt.xlabel('Voltage (V)')
        plt.ylabel('Current (A)')
        plt.show()

    previous_sample_id = sample_id
```
