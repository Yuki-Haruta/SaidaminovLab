```python
import os
import shutil
import pandas as pd
import glob

"""
Transfer csv files from 0_TEMPORAL and empty 0_TEMPORAL
"""

# Specify the directory path
dir_path = '/Users/yukiharuta/Desktop/Temp-Data/MAPbBr3-Paper/02_XRD/0_TEMPORAL'
# Specify the destination directory path
key1, destination_dir_1 = 'gonio', '/Users/yukiharuta/Desktop/Temp-Data/MAPbBr3-Paper/02_XRD/2_GONIO'
key2, destination_dir_2 = '100omega', '/Users/yukiharuta/Desktop/Temp-Data/MAPbBr3-Paper/02_XRD/3_100omega'
key3, destination_dir_3 = '200omega', '/Users/yukiharuta/Desktop/Temp-Data/MAPbBr3-Paper/02_XRD/4_200omega'

destination_dir_others = '/Users/yukiharuta/Desktop/Temp-Data/MAPbBr3-Paper/02_XRD/5_Others'

# Get a list of full paths of folders in the directory
folder_paths = [os.path.join(dir_path, name) for name in os.listdir(dir_path) if os.path.isdir(os.path.join(dir_path, name))]

print('Working...')

for source_dir in folder_paths:   
    # Loop through each file in the source directory
    for file_name in os.listdir(source_dir):
        # Get the full path of the file
        file_path = os.path.join(source_dir, file_name)
        
        if key1 in file_name:
            file_path_check = os.path.join(destination_dir_1, file_name)
            if os.path.exists(file_path_check):
                print(f'{file_path_check} already exists.')
            if not os.path.exists(file_path_check):
                # Read the CSV file into a DataFrame, skipping XX lines of headers to read it properly
                skiprows = 1
                while True:
                    try:
                        data = pd.read_csv(file_path, skiprows=skiprows)
                        x = data['Angle'].to_list()
                        break
                    except:
                        skiprows += 1
                        if skiprows == 100:
                            break
                # Prepare x and y, and normalized y
                y = data[' Intensity'].to_list()
                y_NORM = [n/max(y) for n in y]

                # Create new dataframe
                df = pd.DataFrame({
                    '2theta (degree)': x,
                    'Intensity': y,
                    'NORM': y_NORM
                })

                # Save new csv
                df.to_csv(file_path)

                shutil.move(file_path, destination_dir_1)
                
        elif key2 in file_name:
            file_path_check = os.path.join(destination_dir_2, file_name)
            if os.path.exists(file_path_check):
                print(f'{file_path_check} already exists.')
            if not os.path.exists(file_path_check):
                # Read the CSV file into a DataFrame, skipping XX lines of headers to read it properly
                skiprows = 1
                while True:
                    try:
                        data = pd.read_csv(file_path, skiprows=skiprows)
                        x = data['Angle'].to_list()
                        break
                    except:
                        skiprows += 1
                        if skiprows == 100:
                            break
                # Prepare x and y
                y = data[' Intensity'].to_list()
                # Normilize intensity
                y_NORM = [n/max(y) for n in y]
                index = y.index(max(y))
                x_NORM = [n - x[index] for n in x]

                # Create new dataframe
                df = pd.DataFrame({
                    'Omega (degree)': x,
                    'Intensity': y,
                    'NORM_omega': x_NORM,
                    'NORM': y_NORM
                })

                # Save new csv
                df.to_csv(file_path)
                shutil.move(file_path, destination_dir_2)
                
        elif key3 in file_name:
            file_path_check = os.path.join(destination_dir_3, file_name)
            if os.path.exists(file_path_check):
                print(f'{file_path_check} already exists.')
            if not os.path.exists(file_path_check):
                # Read the CSV file into a DataFrame, skipping XX lines of headers to read it properly
                skiprows = 1
                while True:
                    try:
                        data = pd.read_csv(file_path, skiprows=skiprows)
                        x = data['Angle'].to_list()
                        break
                    except:
                        skiprows += 1
                        if skiprows == 100:
                            break
                # Prepare x and y
                y = data[' Intensity'].to_list()
                # Normilize intensity
                y_NORM = [n/max(y) for n in y]
                index = y.index(max(y))
                x_NORM = [n - x[index] for n in x]

                # Create new dataframe
                df = pd.DataFrame({
                    'Omega (degree)': x,
                    'Intensity': y,
                    'NORM_omega': x_NORM,
                    'NORM': y_NORM
                })

                # Save new csv
                df.to_csv(file_path)
                shutil.move(file_path, destination_dir_3)
        else:
            try:
                # Move the file to the destination directory
                shutil.move(file_path, destination_dir_others)
            except:
                pass
            
# Loop through each subdirectory
for subdir in os.listdir(dir_path):
    # Get the full path of the subdirectory
    subdir_path = os.path.join(dir_path, subdir)
    # Check if the subdirectory is a directory
    if os.path.isdir(subdir_path):
        # Check if the subdirectory is empty
        if not os.listdir(subdir_path):
            # Remove the empty subdirectory
            os.rmdir(subdir_path)
        else:
            # Raise an error if the subdirectory is not empty
            print(f"{subdir_path} is not empty")

print('COMPLETE')
```
