```python
from glob import glob
import TRPL
from pathlib import Path
import pandas as pd

path = '/Users/yukiharuta/Master/05_UVic/01_Project/04_Perovskite-FCG/14_TRPL'
fig_dir = f'{path}/figure'
save_path = f'{path}/summary.xlsx'
directories = TRPL.list_directories(path)

data_path_list = []
for directory in directories:
    path_list = glob(f'{directory}/*.txt')
    data_path_list = data_path_list + path_list
data_path_list.sort()

# Empty dictionary for data frame
dict = {
    'Sample':[],'Measurement':[],'Info':[],'#':[],'Intensity':[],'A1':[],'A2':[],'A3':[],'t1':[],'t2':[],'t3':[],'c':[],'cut time':[],'r2':[],'r2_log':[],'Note':[]
}

for data_path in data_path_list:
    label = Path(data_path).name[:-4]
    measurement_date = Path(data_path).parents[0].name[:6]
    label = measurement_date + '_' + str(Path(data_path).name[:-4])
    print(label)
    try:
        # Get some info from the path
        sample_id = Path(data_path).name.split('_')[0] + '_' + Path(data_path).name.split('_')[1]
        sample_info = Path(data_path).name.split('_')[2]
        trial = Path(data_path).name.split('_')[-1][-5:-4]

        # Read the data
        data = pd.read_csv(data_path,delimiter=',',skiprows=10,header=None)
        X = data[0].to_list() # time (ns)
        Y = data[1].to_list() # count (cts)
        peak_intensity = max(Y)

        cut_time, results, r2, r2_log = TRPL.TRPL_fit(X, Y, label, fig_dir, figsave=True)

    except Exception as e:
        print(e)
        # Get info
        sample_id = Path(data_path).name[:-4]
        sample_info = sample_id
        trial = 1
        # Read the data
        data = pd.read_csv(data_path,delimiter=',',skiprows=10,header=None)
        X = data[0].to_list() # time (ns)
        Y = data[1].to_list() # count (cts)
        peak_intensity = max(Y)
        cut_time, results, r2, r2_log = TRPL.TRPL_fit(X, Y, label, fig_dir, figsave=True)

    dict['Sample'].append(sample_id)
    dict['Measurement'].append(int(measurement_date))
    dict['Info'].append(sample_info)
    dict['#'].append(int(trial))
    dict['Intensity'].append(int(peak_intensity))
    dict['A1'].append(results[0])
    dict['A2'].append(results[1])
    dict['A3'].append(results[2])
    dict['t1'].append(results[3])
    dict['t2'].append(results[4])
    dict['t3'].append(results[5])
    dict['c'].append(results[6])
    dict['cut time'].append(cut_time)
    dict['r2'].append(r2)
    dict['r2_log'].append(r2_log)
    dict['Note'].append('')
    
df = pd.DataFrame(dict)
df.to_excel(save_path)
TRPL.Design_excel(save_path)
```
