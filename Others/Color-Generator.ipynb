```python
'''
Color Palette Generator
'''
original_palette = ["#009392", "#39B1B5", "#9CCB86", "#E9E29C", "#EEB479", "#E88471", "#CF597E"]
# https://www.molecularecologist.com/2020/04/23/simple-tools-for-mastering-color-in-scientific-figures/
num_colors = 6
# recommended to be less than 7

import matplotlib.pyplot as plt
import numpy as np

def interpolate_color(color1, color2, t):
    r1, g1, b1 = int(color1[1:3], 16), int(color1[3:5], 16), int(color1[5:7], 16)
    r2, g2, b2 = int(color2[1:3], 16), int(color2[3:5], 16), int(color2[5:7], 16)
    
    r_interp = int(r1 + t * (r2 - r1))
    g_interp = int(g1 + t * (g2 - g1))
    b_interp = int(b1 + t * (b2 - b1))
    
    return f"#{r_interp:02X}{g_interp:02X}{b_interp:02X}"

def generate_color_palette(palette, num_colors):
    if num_colors < 2:
        raise ValueError("Number of colors must be at least 2")
    
    step_size = 1.0 / (num_colors - 1)
    generated_colors = []
    
    for i in range(num_colors):
        t = i * step_size
        if i == 0:
            generated_colors.append(palette[0])
        elif i == num_colors - 1:
            generated_colors.append(palette[-1])
        else:
            color_index = int(t * (len(palette) - 1))
            color1 = palette[color_index]
            color2 = palette[color_index + 1]
            generated_colors.append(interpolate_color(color1, color2, t))
    
    return generated_colors

def plot_color_palette(color_list):
    size = 1
    plt.figure(figsize=(len(color_list)*size, size))

    for i, color in enumerate(color_list):
        plt.fill_between([i, i + 1], 0, 1, color=color)
        plt.text(i + 0.5, 0.5, color, color='white', va='center', ha='center', fontsize=10)

    plt.xlim(0, len(color_list))
    plt.ylim(0, 1)
    plt.axis('off')
    plt.show()

generated_palette = generate_color_palette(original_palette, num_colors)
print(generated_palette)
plot_color_palette(generated_palette)
```

```python
original_palette = ['#009392', '#CF597E']
num_colors = 4
generated_palette = generate_color_palette(original_palette, num_colors)
print(generated_palette)
plot_color_palette(generated_palette)
```
